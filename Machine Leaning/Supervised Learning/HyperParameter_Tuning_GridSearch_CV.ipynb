```python
import pandas as pd
```

```python
df=pd.read_csv('/content/breast_cancer.csv')
df.head()
```

```python
df.outcome.value_counts()
```

```python
df.shape
```

```python
# x and y split

x=df.iloc[:,:-1]
x.head()
```

```python
y=df.outcome
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=47)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
# Model building

from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred1=model.predict(x_train)
```

```python
# Evalute the model

from sklearn.metrics import accuracy_score, classification_report
```

```python
print("Test Accuracy",accuracy_score(y_test,y_pred))
print("Test Accuracy",accuracy_score(y_train,y_pred1))
```

```python
pd.crosstab(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

## Hyper Parameter Tuning

### GridSearch CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'kernel':['linear','rdf'],
            'C':[0,0.1,0.2,0.5,1.0],
            'gamma':['scale','auto']}
```

```python
clf = GridSearchCV(model,param_grid = parameters, verbose = 2)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_score_
```

```python
clf.best_params_
```

```python
model1=SVC(C= 1.0, gamma='scale', kernel= 'linear')
```

```python
model1.fit(x_train,y_train)
```

```python
y_=model1.predict(x_test)
```

```python
accuracy_score(y_test,y_)
```

```python
pd.crosstab(y_test,y_)
```

```python
print(classification_report(y_test,y_))
```
