```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/car_evaluation.csv')
df.head()
```

```python
df.shape
```

```python
df.outcome.value_counts() # this data is very imbalanced
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn .preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.buying=le.fit_transform(df.buying)
df.maint=le.fit_transform(df.maint)
df.lug_boot=le.fit_transform(df.lug_boot)
df.safety=le.fit_transform(df.safety)
```

```python
df.head()
```

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df.outcome
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.value_counts()
```

```python
# Model building

# KNN-it use for multi cased problems , it is distance based(eucler distance).

from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('Testing Accuracy = ', accuracy_score(y_test,y_pred))
```

```python
pd.crosstab(y_test,y_pred)
```

```python
# testing 1

model.predict([[3,2,2,1,1,0]])
```

```python
#testing 2
model.predict([[1,1,5,6,0,2]])
```

```python
# smote - balancing the data

from imblearn.over_sampling import SMOTE
smote= SMOTE()
```

```python
x_train_smote,y_train_somte=smote.fit_resample(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
y_train_somte.value_counts()  # adding dupicate to make balanced
```

```python
model.fit(x_train_smote,y_train_somte)
```

```python
y_pred1=model.predict(x_test)
```

```python
pd.crosstab(y_test,y_pred1)
```
