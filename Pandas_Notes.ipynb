#Merge

```python
import pandas as pd
```

```python
m1=pd.DataFrame({'name':['manoj','vermeil','krishna','nakamura'],'english':[100,80,100,100]})
m1
```

```python
m2=pd.DataFrame({'name':['manoj','vermeil','krishna','nakamura'],'math':[80,100,80,80]})
m2
```

```python
m=pd.merge(m1,m2, on ='name') # it can merge only two datasets
```

```python
m
```

```python
m3=pd.DataFrame({'name':['manoj','vermeil','krishna','nakamura'],'science':[100,80,100,100]})
m3
```

```python
m4=pd.merge(m,m3, on ='name')
```
