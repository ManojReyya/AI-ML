##K means on 2D-Data

```python
import numpy as np
import pandas as pd
```

```python
ds=pd.read_csv("/content/student_clustering.csv")
ds.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(ds['cgpa'],ds['iq']) # this is data visualization method to find No of clusters
```

```python
# elbow method
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
  km= KMeans(n_clusters=i)
  km.fit_predict(ds)
  wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
# we can see that at 4 we got elbow point, that shows no. of clusters is 4
```

```python
x= ds.iloc[:,:].values
km=KMeans(n_clusters=4)
y_means=km.fit_predict(x)
```

```python
y_means
```

```python
x
```

```python
x[y_means==3,1]
```

```python
plt.scatter(x[y_means==0,0],x[y_means==0,1],color='blue')
plt.scatter(x[y_means==1,0],x[y_means==1,1],color='red')
plt.scatter(x[y_means==2,0],x[y_means==2,1],color='green')
plt.scatter(x[y_means==3,0],x[y_means==3,1],color='yellow')
```

```python
# this is how we can cluster our 2d-data
```

##K means on 3D-Data

```python
from sklearn.datasets import make_blobs

centroids = [(-5,-5,5),(5,5,-5),(3.5,-2.5,4),(-2.5,2.5,-4)]
cluster_std = [1,1,1,1]

X,y = make_blobs(n_samples=200,cluster_std=cluster_std,centers=centroids,n_features=3,random_state=1)
```

```python
import plotly.express as px
fig = px.scatter_3d(x=X[:,0], y=X[:,1], z=X[:,2])
fig.show()
```

```python
wcss = []
for i in range(1,21):
    km = KMeans(n_clusters=i)
    km.fit_predict(X)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,21),wcss)
```

```python
km = KMeans(n_clusters=4)
y_pred = km.fit_predict(X)
```

```python
df = pd.DataFrame()

df['col1'] = X[:,0]
df['col2'] = X[:,1]
df['col3'] = X[:,2]
df['label'] = y_pred
```

```python
fig = px.scatter_3d(df,x='col1', y='col2', z='col3',color='label')
fig.show()
```

##K means clustering for image segmentation

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
img= mpl.image.imread('/content/img.jpg')
img
```

```python
img.shape
```

```python
x=img.reshape(-1,3)
x.shape
```

```python
kmeans=KMeans(n_clusters=4,n_init=10)
```

```python
kmeans.fit(x)
```

```python
seg_img=kmeans.cluster_centers_[kmeans.labels_]
seg_img=seg_img.reshape(img.shape)
```

```python
plt.imshow(seg_img/255)
```
