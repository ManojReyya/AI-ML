### Outliler  Treatment -Replacment outliers with median

```python
#Upper_limit=q3+1.5*IQR

#Lower_limit=q1-1.5*IQR
```

```python
ds.describe()
```

```python
q1=ds.bmi.quantile(0.25)
q1
```

```python
q3=ds.bmi.quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
Upper_limit=q3+1.5*IQR
Upper_limit
```

```python
Lower_limit=q1-1.5*IQR
Lower_limit
```

```python
ds.bmi.median()
```

```python
# replace the outliers with median
```

```python
ds['bmi']=np.where(ds['bmi']>Upper_limit,30,ds['bmi'])
```
